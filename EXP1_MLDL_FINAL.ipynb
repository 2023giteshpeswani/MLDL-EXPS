```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Exercise 1: NumPy Basics
data = pd.read_csv("student_performance.csv")
final_scores = data["Final_Score"].to_numpy()
final_scores
```

```python
mean_score = np.mean(final_scores)
median_score = np.median(final_scores)
std_score = np.std(final_scores)

mean_score, median_score, std_score
```

```python
min_max_norm = (final_scores - np.min(final_scores)) / (np.max(final_scores) - np.min(final_scores))
min_max_norm
```

```python
## Pandas Data Handling
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
def performance_label(score):
    if score >= 75:
        return "Excellent"
    elif score >= 50:
        return "Average"
    else:
        return "Poor"

df["Performance"] = df["Final_Score"].apply(performance_label)
df.head()
```

```python
## Exercise 3: Matplotlib Visualization
plt.figure()
plt.plot(df["Hours_Studied"], df["Final_Score"])
plt.xlabel("Hours Studied")
plt.ylabel("Final Score")
plt.title("Line Plot: Hours Studied vs Final Score")
plt.show()
```

```python
plt.figure()
plt.hist(df["Final_Score"], bins=10)
plt.xlabel("Final Score")
plt.ylabel("Frequency")
plt.title("Histogram Plot: Distribution of Final Scores")
plt.show()
```

```python
## Exercise 4: Seaborn Visualization
plt.figure()
sns.scatterplot(x="Hours_Studied", y="Final_Score", data=df)
plt.title("Scatter Plot: Hours Studied vs Final Score")
plt.show()
```

```python
plt.figure()
corr = df.drop("Performance", axis=1).corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
## Box Plot Categorial Analysis
plt.figure()
sns.boxplot(x="Performance", y="Final_Score", data=df)
plt.title("Final Score Distribution by Performance Category")
plt.show()
```
